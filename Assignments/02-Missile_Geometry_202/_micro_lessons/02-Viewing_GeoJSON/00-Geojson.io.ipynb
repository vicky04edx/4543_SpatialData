## 00 — geojson.io

Before writing a single line of code to render a map, there is a tool you should know about: **[geojson.io](https://geojson.io)**.

It is a free, browser-based editor that lets you create, inspect, edit, and validate GeoJSON visually — no setup required. It is useful at every stage of a project: sketching out features before you code them, debugging malformed files, and quickly checking that coordinates land where you expect.

## The Interface

When you open geojson.io you get two panels side by side:

```
┌─────────────────────────┬──────────────────────────┐
│                         │                          │
│      Interactive        │     JSON editor          │
│         Map             │     (live text)          │
│                         │                          │
│   draw tools ←          │  {                       │
│   on the left edge      │    "type": "Feature...   │
│                         │                          │
└─────────────────────────┴──────────────────────────┘
```

**The two panels are linked.** Draw something on the map and the JSON appears on the right instantly. Edit the JSON on the right and the map updates on the left. You can work from either side.

## Adding Features

The draw toolbar runs along the left edge of the map. Each tool creates a different geometry type:

| Tool | Creates |
|---|---|
| Marker | `Point` |
| Polyline | `LineString` |
| Polygon | `Polygon` |
| Rectangle | `Polygon` (axis-aligned) |

**To add a feature:**
1. Click a draw tool
2. Click on the map to place points
3. Double-click (or click the first point again for polygons) to finish

The new feature appears in the JSON panel immediately as a complete GeoJSON object with `type`, `geometry`, and an empty `properties` block.

## Editing and Moving Features

You do not have to delete and redraw to make a correction.

**To move a feature:** Click the Edit (pencil) tool in the toolbar, then drag the feature to a new location. For lines and polygons, each vertex appears as a handle you can drag individually.

**To edit properties:** Click any feature on the map. A popup appears with a small table of its `properties`. You can add, rename, or change values directly in the table — no need to touch the JSON.

**To delete a feature:** Select the trash tool, click the feature, then click "Save".

Every one of these actions updates the JSON panel in real time.

## Linting — Checking for Errors

The JSON panel is not just a viewer — it actively validates your GeoJSON as you type.

**What it catches:**
- Syntax errors (missing brackets, trailing commas, unquoted keys)
- Invalid `type` values
- Coordinates in the wrong order or wrong nesting level — a common mistake with Polygons

**How to use it to lint a file you already have:**
1. Open your `.geojson` file in a text editor and copy the contents
2. Paste it into the right panel at geojson.io
3. If the map renders correctly, the file is valid
4. If the map is blank or shows an error banner, the JSON panel will highlight the problem line

This is faster than debugging a `FileNotFoundError` or a blank map in a notebook. **When something looks wrong, paste it into geojson.io first.**

## Importing and Exporting

geojson.io is not just for drawing from scratch — you can load existing files.

**Import:** `Open` menu → upload a local `.geojson`, `.json`, or `.csv` file. The features load onto the map immediately.

**Export:** `Save` menu → download as GeoJSON, CSV, KML, or SVG.

The most common workflow is:
1. Draw or edit features in geojson.io
2. Download as GeoJSON
3. Drop the file into your project's `data/` folder
4. Load it in Python with `json.loads()`
