## 01 Relative vs Absolute

## Goal

- Files can be described using **absolute paths** or **relative paths**.
- You need to understand both, but should lean toward **relative paths** in real projects.

## Absolute Path

- An absolute path begins from the root of the machine.

- **Examples:**
  - `/Users/terry/project/data/file.json`
  - `C:/Users/Student/project/data/file.json`

- Absolute paths are specific to one machine and one layout.
- It always starts with a `/` (slash) or a drive letter (`C:`). 

## Relative Path

- A relative path begins from the **working directory**.

**Example:**
  - `data/file.json`

- Relative paths make projects portable.
- They never start with a slash (`/`) unless it is an explicit "here" by using a "dot slash" (e.g. `./data/file.json`)
- So: `data/file.json` == `./data/file.json`

```python
from pathlib import Path

absolute_example = Path("/")
relative_example = Path("data/file.json")

print("Absolute path example:", absolute_example)
print("Relative path example:", relative_example)
print("Resolved relative path from cwd:", Path.cwd() / relative_example)
```

## Key Rule

In course projects and professional code, prefer **relative paths** whenever possible.

That way the same notebook can run on your machine, your student's machine, or a lab machine without path surgery.

```python
examples = [
    Path("data/countries.geojson"),
    Path("../data/countries.geojson"),
    Path("./notes.txt"),
]

for p in examples:
    print(f"{p!s:<30} -> {(Path.cwd() / p).resolve()}")
```

## Optional Advanced — `resolve()` vs `cwd() / p`

Both `Path.cwd() / p` and `p.resolve()` can turn a relative path into an absolute one.

Run the cell below and compare the two. Are they always the same? When might they differ?

```python
from pathlib import Path

# Try both methods on the same relative path with ".." components
p = Path("../data/countries.geojson")

# Method 1: cwd() / p
method1 = Path.cwd() / p

# Method 2: resolve()
method2 = p.resolve()

print("cwd() / p  :", method1)
print("resolve()  :", method2)
print("Same result?", method1.resolve() == method2)
```
