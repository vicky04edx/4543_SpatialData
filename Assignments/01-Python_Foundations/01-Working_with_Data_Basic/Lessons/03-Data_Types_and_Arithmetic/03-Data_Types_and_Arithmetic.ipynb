## ✅ Learning Goals
- Understand Python data types
- Use variables and print statements
- Perform basic arithmetic

```python
# 🧠 Variables and Types
x = 10
y = 3.5
z = 'hello'
print(type(x), type(y), type(z))
```

```python
# ➕ Arithmetic Operations
print('Add:', x + y)
print('Divide:', x / y)
print('Floor Divide:', x // y)
print('Exponent:', x ** 2)
```

## ✏️ Try It Yourself
- Create your own variables
- Try combining numbers and strings
- Use `input()` to interact

```python
name = input('What is your name? ')
age = int(input('How old are you? '))

year = 2026
birthYear = year - age

print('Welcome ' + name)
print('You are ' + str(age) + ' years old')
print('You were probably born in ' + str(birthYear))
name = input('What is your name? ')
print(name + ' was born in ' + str(birthYear) + ' and is ' + str(age) + ' years old')
```
