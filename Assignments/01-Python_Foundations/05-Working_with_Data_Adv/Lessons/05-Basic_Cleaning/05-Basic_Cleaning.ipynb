# 🧼 Cleaning Dirty Data (Missing Values & Type Fixes)

## 🔹 LEARNING GOALS:
- Detect and count missing values (`NaN`)
- Fill or drop missing data
- Convert column data types safely
- Understand the difference between `NaN`, `None`, `""`, and type mismatches


### 🧪 1. Load a Messy Dataset

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": ["25", "thirty", 35, np.nan, "40"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022"],
    "Score": [95.5, None, 88.0, 92.5, ""]
}

df = pd.DataFrame(data)
df
```

### 🧯 2. Detecting Missing or Broken Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

### 🧹 3. Cleaning Strategy Options

```python
df.fillna({
    "Name": "Unknown",
    "Age": -1,
    "Signup Date": "1970-01-01",
    "Score": 0.0
})
```

```python
df.dropna()
```

### 🧬 4. Data Type Fixes

```python
df.dtypes
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df.dtypes
```

### 🩹 5. Impute (Fill In) Fixed Missing Values

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Score"].fillna(df["Score"].mean(), inplace=True)
```

```python
df["Signup Date"].fillna(df["Signup Date"].min(), inplace=True)
```

```python
df["Name"].fillna("Unknown", inplace=True)
```

### 🤓 6. Cleaned Data Review

```python
df.info()
```

```python
df.describe(include="all")
```

### 🧪 Try It Yourself

Modify the `data` dictionary at the top of this notebook. Add:
- A new column with some `None` and `""` values
- At least one row with all columns filled incorrectly
Then re-run the notebook and fix it step-by-step.

```python
import pandas as pd

# Modified data (with issues intentionally added)
data = {
    "Name": ["Alex", "Jordan", "Taylor", "Sam", ""],   # empty string
    "GPA": [3.8, None, 3.9, 3.7, "oops"],            # None + wrong type
    "Credits": [90, 75, None, 60, -10],              # None + invalid value
    "Graduating": [True, False, True, None, "yes"],  # None + wrong type
    "Notes": ["Good", "", None, "Late", "???"]       # new column with None + ""
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)


df["GPA"] = pd.to_numeric(df["GPA"], errors="coerce")  # fix "oops"
df["Credits"] = pd.to_numeric(df["Credits"], errors="coerce")

df.replace("", pd.NA, inplace=True)

df = df[df["Credits"] >= 0]


df["Graduating"] = df["Graduating"].replace({"yes": True})
df["Graduating"] = df["Graduating"].fillna(False)

df = df.dropna(subset=["Name", "GPA"])  # remove critical missing values

print("\nCleaned Data:")
print(df)
```

### 🧠 Mini-Challenge

> 🗂 Load `"data/survey.csv"` and:
- Identify which columns have missing values
- Use `.isnull().sum()` to get a null report
- Use a mix of `.fillna()`, `.dropna()`, and `pd.to_numeric()` or `pd.to_datetime()` to clean it
- Print a summary with `.info()` and `.describe()`

```python
import pandas as pd
import os

# Check if file exists, otherwise create sample data
if os.path.exists("data/survey.csv"):
    df = pd.read_csv("data/survey.csv")
else:
    # Create sample survey data
    df = pd.DataFrame({
        "Age": [25, None, 35, 40, "invalid", 28],
        "Satisfaction": [8, 9, None, 7, 6, ""],
        "Income": [50000, 60000, None, 75000, "", 55000],
        "date": ["2023-01-15", "2023-02-20", None, "2023-04-10", "invalid", "2023-06-05"]
    })

print("Missing values per column:")
print(df.isnull().sum())


# Convert numeric columns
for col in df.columns:
    if col != "date":
        df[col] = pd.to_numeric(df[col], errors="coerce")

if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

for col in df.select_dtypes(include=["number"]).columns:
    df[col].fillna(df[col].mean(), inplace=True)

for col in df.select_dtypes(include=["object"]).columns:
    df[col].fillna("Unknown", inplace=True)

df = df.dropna()

print("\nDataFrame Info:")
df.info()

print("\nSummary Statistics:")
print(df.describe())
```
