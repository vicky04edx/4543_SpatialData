## 📥 Load a CSV

Replace the filename below with a real CSV path or URL. For testing, use built-in seaborn datasets.

````python
# Example with seaborn's Titanic dataset
import seaborn as sns
df = sns.load_dataset("titanic")
df.head()
````

## 📊 Peek at the Data

````python
# First and last few rows
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())
````

## 🧠 Understand Structure

````python
# Dimensions and columns
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
````

## 🧼 Data Types & Missing Values

````python
# Data types and null counts
print("\nInfo:")
df.info()

# How many missing values per column?
print("\nMissing values per column:")
print(df.isnull().sum())
````

## 📈 Summary Statistics

````python
df.describe(include="all")
````

## ✏️ Renaming Columns (Optional but fun)

````python
# Rename 'sex' to 'gender' for clarity
df.rename(columns={"sex": "gender"}, inplace=True)
df.head(2)
````

````python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("tips")

print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nDataFrame Info:")
df.info()

print("\nSummary Statistics:")
print(df.describe())

print("\nColumns before rename:")
print(df.columns)

df.rename(columns={"total_bill": "total_bill_usd"}, inplace=True)

print("\nColumns after rename:")
print(df.columns)

percent_missing = df.isnull().any(axis=1).mean() * 100
print(f"\nPercentage of rows with missing values: {percent_missing:.2f}%")
````
